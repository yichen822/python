```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from math import pi

sns.set(style="whitegrid")

df = pd.read_excel('SHOP_DATA_ADVANCED_220_ROWS.xlsx')

num_cols = df.select_dtypes(include='number').columns
cat_cols = df.select_dtypes(include='object').columns

plt.figure(figsize=(50, 50))
plot_idx = 1
```

```python
# 1 Bar
if len(cat_cols) and len(num_cols):
    plt.subplot(5,4,plot_idx)
    df.groupby(cat_cols[0])[num_cols[0]].sum().plot(kind='bar')
    plt.title("Bar"); plot_idx+=1
```

```python
# 2 Line
if len(num_cols):
    plt.subplot(5,4,plot_idx)
    df[num_cols].plot(ax=plt.gca())
    plt.title("Line"); plot_idx+=1
```

```python
# 3 Pie
if len(cat_cols):
    plt.subplot(5,4,plot_idx)
    df[cat_cols[0]].value_counts().plot.pie(autopct='%1.1f%%')
    plt.title("Pie"); plot_idx+=1
```

```python
# 4 Scatter
if len(num_cols)>=2:
    plt.subplot(5,4,plot_idx)
    sns.scatterplot(x=num_cols[0], y=num_cols[1], data=df)
    plt.title("Scatter"); plot_idx+=1
```

```python
# 5 Stacked Bar
if len(cat_cols)>=2 and len(num_cols):
    plt.subplot(5,4,plot_idx)
    df.pivot_table(index=cat_cols[0], columns=cat_cols[1], values=num_cols[0], aggfunc='sum').plot(kind='bar', stacked=True, ax=plt.gca())
    plt.title("Stacked Bar"); plot_idx+=1
```

```python
# 6 Box
if len(num_cols):
    plt.subplot(5,4,plot_idx)
    sns.boxplot(data=df[num_cols])
    plt.title("Box"); plot_idx+=1
```
